Association-Apriori Algorithm

```python
# Import necessary libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Load the transaction dataset
data = pd.read_csv('/content/drive/MyDrive/Machine_Learning_Techniques/Instacart market Basket Dataset.csv')
data
```

```python
data = data.drop(columns=['Item(s)'])
```

```python
# Convert data into list of transactions
transactions = []
for i in range(len(data)):
    transactions.append([str(item) for item in data.iloc[i] if str(item) != 'nan'])
```

```python
# Convert into One-Hot Encoding (required for Apriori)
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_data, columns=te.columns_)
```

```python
print("(One-Hot Encoded):")
print(df.head())
```

```python
# Apply Apriori algorithm
min_support = 0.05
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)
print("Frequent Itemsets:")
print(frequent_itemsets)

# Generate Association Rules
min_confidence = 0.5
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
print("Association Rules:")
print(rules)
```

```python
print(frequent_itemsets.sort_values(by='support', ascending=False).head(5))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Top 10 Frequent Itemsets
top10 = frequent_itemsets.sort_values(by='support', ascending=False).head(10)

# Convert itemsets to strings for labels
top10['Itemsets_str'] = top10['itemsets'].apply(lambda x: ', '.join(list(x)))

plt.figure(figsize=(12,6))
plt.barh(top10['Itemsets_str'], top10['support'], color='skyblue')
plt.xlabel('Support')
plt.ylabel('Itemsets')
plt.title('Top 10 Frequent Itemsets (Apriori)')
plt.gca().invert_yaxis()  # Highest support on top
plt.show()
```

```python
#Heatmap
te_data_df = pd.DataFrame(te_data, columns=te.columns_)
# Compute co-occurrence matrix for top 10 items
top_items = top10['Itemsets_str'].apply(lambda x: x.split(', '))
# Flatten to get unique items
top_items_flat = sorted({item.strip() for sublist in top_items for item in sublist})
co_matrix = te_data_df[top_items_flat].T.dot(te_data_df[top_items_flat])
plt.figure(figsize=(8,6))
sns.heatmap(co_matrix, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Top Item Co-Occurrences')
plt.show()
```

```python
#Scatter Plot: Itemset Length vs Support
frequent_itemsets['Length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
plt.figure(figsize=(10,6))
plt.scatter(frequent_itemsets['Length'], frequent_itemsets['support'], alpha=0.6, color='purple')
plt.xlabel('Itemset Length')
plt.ylabel('Support')
plt.title('Scatter Plot: Itemset Length vs Support')
plt.grid(True)
plt.show()
```

Association-Eclat Algorithm

```python
# Import library
import pandas as pd
```

```python
# Load the transaction dataset
data = pd.read_csv('/content/drive/MyDrive/Machine_Learning_Techniques/Instacart market Basket Dataset.csv')
data
```

```python
data = data.drop(columns=['Item(s)'])
```

```python
# Convert into transactions
transactions = []
for i in range(len(data)):
    transactions.append([str(item) for item in data.iloc[i] if str(item) != 'nan'])
```

```python
# Create vertical format (item → transaction IDs)
item_tid = {}
for tid, transaction in enumerate(transactions):
    for item in transaction:
        if item not in item_tid:
            item_tid[item] = set()
        item_tid[item].add(tid)
```

```python
# Minimum support (count)
min_support = 2
```

```python
# ECLAT function
def eclat(prefix, items, min_support, freq_itemsets):
    while items:
        item, tidset = items.pop()
        support = len(tidset)

        if support >= min_support:
            new_itemset = prefix + [item]
            freq_itemsets.append((new_itemset, support))

            new_items = []
            for other_item, other_tidset in items:
                intersection = tidset & other_tidset
                if len(intersection) >= min_support:
                    new_items.append((other_item, intersection))

            eclat(new_itemset, new_items, min_support, freq_itemsets)
```

```python

# Prepare items
items = list(item_tid.items())
```

```python
# Store results
frequent_itemsets = []
```

```python
# Run ECLAT
eclat([], items, min_support, frequent_itemsets)
```

```python
# Convert to DataFrame
result = pd.DataFrame(frequent_itemsets, columns=['Itemsets', 'Support'])
```

```python
# Show top results
print("Frequent Itemsets (ECLAT):")
print(result.sort_values(by='Support', ascending=False).head())
```

```python
# Visualization: Bar Chart
import matplotlib.pyplot as plt

# Sort by Support
result_sorted = result.sort_values(by='Support', ascending=False)

# Take top 10 frequent itemsets
top10 = result_sorted.head(10)

# Convert itemsets to string for labels
top10['Itemsets_str'] = top10['Itemsets'].apply(lambda x: ', '.join(x))

# Plot horizontal bar chart
plt.figure(figsize=(12,6))
plt.barh(top10['Itemsets_str'], top10['Support'], color='skyblue')
plt.xlabel('Support (Count)')
plt.ylabel('Itemsets')
plt.title('Top 10 Frequent Itemsets (ECLAT)')
plt.gca().invert_yaxis()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Convert transactions to a DataFrame of one-hot encoded items
all_items = sorted({item for t in transactions for item in t})
onehot = pd.DataFrame([{item: (item in t) for item in all_items} for t in transactions])

# Compute co-occurrence matrix
co_matrix = onehot.T.dot(onehot)

# Take top 10 frequent items for readability
top_items = result.sort_values(by='Support', ascending=False).head(50)['Itemsets']
top_items_flat = sorted({i for sublist in top_items for i in sublist})
co_matrix_top = co_matrix.loc[top_items_flat, top_items_flat]

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(co_matrix_top, annot=True, fmt='d', cmap='YlGnBu')
plt.title("Heatmap of Top Item Co-Occurrences")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Add a column for itemset length
result['Length'] = result['Itemsets'].apply(len)

# Scatter plot: Itemset Length vs Support
plt.figure(figsize=(10,6))
plt.scatter(result['Length'], result['Support'], color='purple', alpha=0.6)
plt.xlabel('Itemset Length')
plt.ylabel('Support (Count)')
plt.title('Scatter Plot: Itemset Length vs Support')
plt.grid(True)
plt.show()
```
